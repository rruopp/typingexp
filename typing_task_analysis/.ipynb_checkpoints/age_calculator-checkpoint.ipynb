```python
from datetime import datetime, date

# Example input
birth_str = '07/12/06'
study_date_str = '5/27/25'  # e.g., date of experiment

# Convert to datetime.date objects
birth_date = datetime.strptime(birth_str, '%m/%d/%y').date()
study_date = datetime.strptime(study_date_str, '%m/%d/%y').date()

# Calculate age at time of study
age = study_date.year - birth_date.year - (
    (study_date.month, study_date.day) < (birth_date.month, birth_date.day)
)

age
```
